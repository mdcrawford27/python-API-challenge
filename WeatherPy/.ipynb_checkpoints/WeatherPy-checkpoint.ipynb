```python
# dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress

# import api key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# output file
output_data_file = "output_data/cities.csv"

# range of latitudes and longitudes
lat_range = (-90,90)
lng_range = (-180,180)
```

```python
# create lists for holding lat/lngs and cities
lat_lngs = []
cities = []

# create a random set of latitude and longitude coordinates
lats = np.random.uniform(lat_range[0], lat_range[1], size = 1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size = 1500)
lat_lngs = zip(lats,lngs)

# identifying nearest city for each coordinate
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # if the city is unique add it to our list of cities
    if city not in cities:
        cities.append(city)

#print the city count to confirm at least 500 cities
len(cities)
```

```python
# url for OpenWeather API call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key + "&q="

# create list for city's weather info
city_data = []

# create "logger"
print("Beginning Data Retrieval")
print("----------------------------------------")

# create counter
record_count = 1

# loop through all the cities in our list
for city in cities:
    
    # add city to url
    city_url = url + city
    
    # print the record to the logger
    print(f'Processing Record {record_count} | {city}')
    
    # add one to record count
    record_count = record_count + 1
    
    # run the API request for each of the cities
    try:
        # retrieve json data
        city_weather = requests.get(city_url).json()
        
        # identify specific weather data that is needed
        city_lat = city_weather["coord"]["lat"] 
        city_lng = city_weather["coord"]["lon"]
        city_country = city_weather["sys"]["country"]
        city_date = city_weather["dt"]
        city_max = city_weather["main"]["temp_max"]
        city_hum = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        
        # add the data above into the city_data list
        city_data.append({"City": city,
                        "Latitude": city_lat,
                        "Longitude": city_lng,
                        "Max Temp": city_max,
                        "Humidity": city_hum,
                         "Cloudiness": city_clouds,
                         "Wind Speed": city_wind,
                         "Country": city_country,
                         "Date": city_date})
    
    # to skip the city if there is an error in data
    except:
        print("City not found. Skipping...")
        pass
    
# to complete the logger
print("-----------------------------------------")
print("Data Retrieval is Complete")
print("-----------------------------------------")
```

```python
# convert data pulled from API into datafram
city_data_pd = pd.DataFrame(city_data)

# count data points
city_data_pd.count()
```

```python
# create a function to plot linear regressions
def plot_LR(x_values, y_values, y_label, grid):
    
    # run regression
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    # plot scatter and linear regression
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,grid,fontsize=15,color="red")
    plt.xlabel("Latitude")
    plt.ylabel(y_label)
    print(f"The r-squared is: {rvalue**2}")
    plt.show()
```

```python
# create Northern and Southern Hemisphere data frames
north_hem = city_data_pd.loc[(city_data_pd["Latitude"] >= 0)]
south_hem = city_data_pd.loc[(city_data_pd["Latitude"] < 0)]
```

```python
# LATITUDE VS MAX TEMP LINEAR REGRESSION

# northern hemisphere
x_values = north_hem["Latitude"]
y_values = north_hem["Max Temp"]
plot_LR(x_values, y_values, "Max Temp", (0,40))
```

```python
# southern hemisphere
x_values = south_hem["Latitude"]
y_values = south_hem["Max Temp"]
plot_LR(x_values, y_values, "Max Temp", (-25,32))
```

```python
# The r-values for both hemispheres indicate a positive correlation between latitude and temperature, with the correlation
# in the southern hemisphere being a little stronger.
```

```python
#  LATITUDE VS HUMIDITY LINEAR REGRESSION

# northern hemisphere
plot_LR(north_hem["Latitude"],north_hem["Humidity"],"Humidity",(45,7))
```

```python
# southern hemisphere
plot_LR(south_hem["Latitude"],south_hem["Humidity"],"Humidity",(-57,16))
```

```python
# The low r values indicate very little/zero correlation between latitude and humidity.
```

```python
# LATITUDE VS CLOUDINESS LINEAR REGRESSION

# northern hemisphere
plot_LR(north_hem["Latitude"],north_hem["Cloudiness"],"Cloudiness",(-2,5))
```

```python
# southern hemisphere
plot_LR(south_hem["Latitude"],south_hem["Cloudiness"],"Cloudiness",(-55,15))
```

```python
# The low r values indicate very little/zero correlation between latitude and cloudiness.
```

```python
# LATITUDE VS WIND SPEED LINEAR REGRESSION

# northern hemisphere
plot_LR(north_hem["Latitude"],north_hem["Wind Speed"],"Wind Speed",(40,25))
```

```python
# southern hemisphere
plot_LR(south_hem["Latitude"],south_hem["Wind Speed"],"Wind Speed",(-55,22))
```
