```python
# dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
```

```python
# import API key
from api_keys import g_key
```

```python
# create dataframe based off csv created in the first part
city_data = pd.read_csv("output_data/cities.csv")
city_data.head()
```

```python
# configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# humidity heatmap

locations = city_data[["Latitude", "Longitude"]]
humidity = city_data["Humidity"]
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=150,
                                 point_radius = 3)

fig.add_layer(heat_layer)

fig
```

```python
# narrow down cities to fit your ideal weather condition

ideal_cities = city_data.loc[(city_data["Max Temp"] < 85) & (city_data["Max Temp"] > 68) \
                             & (city_data["Wind Speed"] < 12) & (city_data["Cloudiness"] < 10)].dropna()

ideal_cities
```

```python
# create hotel dataframe
hotel_df = ideal_cities[["City","Country","Latitude","Longitude"]].copy()

# add hotel name column
hotel_df["Hotel Name"] = ""

hotel_df
```

```python
# set up parameters to search for hotels within 5000 meters

params = {"types":"lodging", "radius":5000, "key":g_key}

# use iterrows to find hotels based on latitude and longitude
for index, row in hotel_df.iterrows():
    
    # pull lat and lng
    lat = row["Latitude"]
    lng = row["Longitude"]
    
    # add location to our params
    params["location"] = f"{lat},{lng}"
    
    # url from Google API
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    # make request/convert to json
    name_hotel = requests.get(url, params=params).json()
    
    # grab the first hotel and store in dataframe
    try:
        hotel_df.loc[index, "Hotel Name"] = name_hotel["results"][0]["name"]
    except (KeyError, IndexError):
        print("No results...skipping city.")

hotel_df
    
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""

# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
# add marker layer for hotels on top of heat map

marker_layer = gmaps.marker_layer(locations,info_box_content=hotel_info)
fig.add_layer(marker_layer)

fig
```
